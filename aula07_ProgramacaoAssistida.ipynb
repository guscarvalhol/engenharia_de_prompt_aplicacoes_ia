```python
# Define a list of numbers
numeros = [1,3,6,7,8]

# Initialize a variable to store the sum
soma = 0

# Iterate through each number in the list
for num in numeros:
  # Add the current number to the sum
  soma += num

# Print the final sum
print(soma)
```
