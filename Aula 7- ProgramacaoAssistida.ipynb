```python
# Define a list of numbers
numeros = [1,3,6,7,8]

# Initialize a variable to store the sum
soma = 0

# Iterate through each number in the list
for num in numeros:
  # Add the current number to the sum
  soma += num

# Print the final sum
print(soma)
```

Mini Projeto

```python
# Mini projeto: Calculadora

def calculadora():
    a = 10.5
    b = 2.0

    print("=== Calculadora em python ===")
    print("Valores usados:", a, "e", b)

    print("\nSoma:", a + b)
    print("Subtração:", a - b)
    print("Multiplicação:", a * b)

    if b != 0:
        print("Divisão:", a / b)
    else:
        print("Erro: divisão por zero")

# Executa
calculadora()
```

Inversão de String

```python
def inverte_texto(texto):
  # Esta função recebe uma string e a retorna invertida
  # Utilizamos o fatiamento (slicing) [::-1] que é a forma mais eficiente em Python
  return texto[::-1]

# Exemplo de uso:
nome = "gustavo"
resultado = inverte_texto(nome)

# Exibe o resultado da inversão
print(f"Original: {nome}")
print(f"Invertido: {resultado}")
```

Eh primo

```python
# Verificador de número primo

def eh_primo(n):
    if n <= 1:
        return False

    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False

    return True

# Número definido
numero = 19

print("Número analisado:", numero)

if eh_primo(numero):
    print("Resultado: É primo")
else:
    print("Resultado: Não é primo")
```
